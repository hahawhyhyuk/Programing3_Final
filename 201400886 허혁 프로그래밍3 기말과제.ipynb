```python
#가입여부와 데이터요금제사용여부를 Groupby하여 평균요금과 초과요금의 최대, 평균, 최소를 봅니다.

(telco.groupby(['Leave', 'IsData'])['MonthlyFee', 'OverFee'].agg(['max', 'mean', 'min']))
```

```python
#메서드체인

def flatten_cols(df):
    df.columns = ['_'.join(x) for x in df.columns.to_flat_index()]
    return df
res = (telco
.groupby(['Leave', 'IsData'])
.agg({'Callcenter':['sum', 'mean', 'size'],
'RoamingMin':['mean', 'var']})
.pipe(flatten_cols)
)
```

```python
res
```

```python
#agg사용을 위해 커스텀 함수를 사용합니다.

def max_deviation(s):
    std_score = (s - s.mean()) / s.std()
    return std_score.abs().max()
```

```python
#가입해지한사람중 데이터 사용량 표준편차를 봅니다.

(telco.groupby('Leave')['DataAmount'].agg(max_deviation))
```

```python
#고객센터에 3번 이상 전화 하는 사람의 비율을 반환합니다

def call_between_2_9(s):
    return (
        s
        .between(2, 9)
        .mean() * 100
    )
```

```python
(telco.groupby(['Leave', 'Callcenter'])['CalMins'].agg(call_between_2_9))
```

```python
#UNSTACK을 사용하여 기준이 되는 값을 입력해줍니다.

(telco.groupby(['Callcenter', 'Leave'])['MonthlyFee'].mean().unstack('Leave'))
```

```python
#반대로 Callcenter로 UNSTACK해줍니다.

(telco.groupby(['Callcenter', 'Leave'])['MonthlyFee'].mean().unstack('Callcenter'))
```

```python
#콜센터 전화 횟수, 초과요금, 월평균요금을 피벗 테이블로 나타내어봤습니다.

pivot = telco.pivot_table(index='Callcenter',
                         columns='OverFee',
                         values='MonthlyFee',
                         aggfunc='sum',
                         fill_value=0)
pivot
```
